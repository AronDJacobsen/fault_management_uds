```python
# Define the transformations as functions
def log_transform(data):
    return np.log1p(data)  # log(x + 1) to avoid issues with zeros

def sqrt_transform(data):
    return np.sqrt(data)

def boxcox_transform(data):
    # Box-Cox requires positive values; shift data if necessary
    return boxcox(data + 1e-6)[0] if np.all(data > 0) else boxcox(data - np.min(data) + 1e-6)[0]

def quantile_transform(data, output_distribution='uniform'):
    qt = QuantileTransformer(output_distribution=output_distribution, random_state=0)
    return qt.fit_transform(data.values.reshape(-1, 1)).flatten()
```

```python
# Iterate through each sensor in natural_sensor_order
for sensor in natural_sensor_order:
    # Extract the data for the current sensor, dropping NaN values
    sensor_data = data[sensor].dropna().copy()

    # List of transformations
    transformations = {
        "Original": sensor_data,
        "Log": log_transform(sensor_data),
        "Square Root": sqrt_transform(sensor_data),
        "Box-Cox": boxcox_transform(sensor_data),
        "Quantile (Uniform)": quantile_transform(sensor_data, output_distribution="uniform"),
    }

    # Visualize the histograms for each transformation
    fig, axes = plt.subplots(1, len(transformations), figsize=(20, 5), dpi=100)

    for ax, (name, transformed_data) in zip(axes, transformations.items()):
        ax.hist(transformed_data, bins=100, color='#4682b4', alpha=0.7, edgecolor='white')
        #ax.set_yscale('log')  # Use log scale for y-axis
        ax.set_title(f"{name} Transformation")
        ax.set_xlabel(f"{sensor}")
        ax.set_xlim(0, 1.1 * np.max(transformed_data))
        ax.set_ylabel("Frequency (log scale)")

    plt.tight_layout()
    plt.show()
    plt.close()
    break  # Remove this if you want to process all sensors
```

```python
# resample some values 

rain_event_priority = 50

# load
rain_events = pd.read_csv(REFERENCE_DIR / 'events' / 'rain_events.csv')
rain_events['start'], rain_events['end'] = pd.to_datetime(rain_events['start']), pd.to_datetime(rain_events['end'])
# go sequence length backwards and forwards
rain_events['start'] = rain_events['start'] #- pd.Timedelta(minutes=180)
rain_events['end'] = rain_events['end'] + pd.Timedelta(minutes=180)
# merge 
rain_events = merge_intervals(rain_events)

# Generate the complete date range
complete_range = pd.concat([
    pd.Series(pd.date_range(row['start'], row['end']))
    for _, row in rain_events.iterrows()
])

# inject priority into the data
#rain_event_priority = dataset_args.get('rain_event_priority', 1)
data['priority_weight'] = 1.0 # default
data.loc[data.index.isin(complete_range), 'priority_weight'] = rain_event_priority

```

```python
# Iterate through each sensor in natural_sensor_order
for sensor in natural_sensor_order:
    sensor = 'G80F11B_Level1'
    # Extract the data for the current sensor, dropping NaN values
    sensor_data = data[sensor].dropna().copy()
    obvious_min = obvious_min_dict[sensor]

    print(sensor_data.shape)    


    # Repeat rows based on 'priority_weight'
    repeated_data = data[[sensor, 'priority_weight']].copy()
    print(repeated_data.shape)
    repeated_data = repeated_data.loc[repeated_data.index.repeat(repeated_data['priority_weight'])].reset_index(drop=True)
    # Select only the 'sensor' column
    repeated_data = repeated_data[sensor].dropna().copy()
    print(repeated_data.shape)


    # List of transformations
    transformations = {
        "Original": sensor_data,
        
        "Min-Max": CustomScaler(scaler_type='min-max', feature_range=(0, 1), function_transform_type=None, obvious_min=obvious_min, ).fit_transform(sensor_data.values.reshape(-1, 1)).flatten(),
        "Log + Min-Max": CustomScaler(scaler_type='min-max', feature_range=(0, 1), function_transform_type='log', obvious_min=obvious_min).fit_transform(sensor_data.values.reshape(-1, 1)).flatten(),
        "Sqrt + Min-Max": CustomScaler(scaler_type='min-max', feature_range=(0, 1), function_transform_type='sqrt', obvious_min=obvious_min).fit_transform(sensor_data.values.reshape(-1, 1)).flatten(),
        "Sqrt + Min-Max + Weighted": CustomScaler(scaler_type='min-max', feature_range=(0, 1), function_transform_type='sqrt', obvious_min=obvious_min).fit_transform(repeated_data.values.reshape(-1, 1)).flatten(),

    }

    # Visualize the histograms for each transformation
    fig, axes = plt.subplots(len(transformations), 1, figsize=(14, 7))

    for ax, (name, transformed_data) in zip(axes, transformations.items()):
        # Calculate the total number of data points
        total_count = len(transformed_data)
        print(name, transformed_data.shape)

        # Plot the histogram with percentage
        ax.hist(transformed_data, bins=150, color='#4682b4', alpha=0.7, edgecolor='white', 
                weights=np.ones_like(transformed_data) * 100 / total_count)

        ax.set_title(f"{name} Transformation")
        #ax.set_xlabel(f"{sensor}")
        ax.set_xlim(np.min(transformed_data), np.max(transformed_data))
        #ax.set_ylabel("%")
    # set a main y label
    fig.text(-0.01, 0.5, 'Percentage (%)', va='center', rotation='vertical', fontsize=16)
    plt.tight_layout()
    plt.show()
    plt.close()
    break  # Remove this if you want to process all sensors
```
